# How Did Covid-19 Influence Gaming Companies' Stock Price?

# Overview

The objective of our project is to analyze the relationship between stock prices for gaming companies and Covid cases from 2020 to 2022. We are looking for if there is any positive or negative correlation between these two data. We choose the 7 most popular gaming companies, such as Nintendo, Microsoft, EA, etc as our dependent variables. We use Covid-cases as our single predictor. We used polynomial regression to analyze the correlation between the predictor and dependent variables.


<a id='research_question'></a>
# Research Question

From 2020 to 2022, what’s the relationship between COVID-19 cases and the stock price of gaming companies in the US? We want to know the stock prices of gaming companies are influenced by the increasing number of COVID-19 cases or not. We decide to use the close prices of stock for each day as the dependent variables and positive COVID-19 cases of each day as independent variables.

# Data Cleaning

First we convert 'Date' column inside each stock dataframe of the gaming companies into datetime object. And then we filter the rows of dataframe, so that it contains the stock prices from 2020-01-22 to 20222-05-04. We then reset the index. Next, we filter columns of covid dataframe. We keep the date columns from 2020-01-22 to 2022-0504 only. We transpose the data and sum of the cases within a same day and assign a new column to it. We drop the unnecessary columns and keep only date column and cases column only. Last, we convert the date column of covid data into datetime object for future merging.Describe your data cleaning steps here.

```python
#Convert each date col in the stock df into date time object
nintendo['Date'] = pd.to_datetime(nintendo['Date'])
act_bliz['Date'] = pd.to_datetime(act_bliz['Date'])
ea['Date'] = pd.to_datetime(ea['Date'])
microsoft['Date'] = pd.to_datetime(microsoft['Date'])
take_two['Date'] = pd.to_datetime(take_two['Date'])
ubisoft['Date'] = pd.to_datetime(ubisoft['Date'])
zynga['Date'] = pd.to_datetime(zynga['Date'])
```

```python
#Filter rows of each stock df to keep the rows from 2020-01-22 to 2022-04-29 only, and reset the index
nintendo_latest = nintendo[nintendo['Date'].dt.year >= 2020]
nintendo_latest = nintendo_latest[nintendo_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
act_bliz_latest = act_bliz[act_bliz['Date'].dt.year >= 2020]
act_bliz_latest = act_bliz_latest[act_bliz_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
ea_latest = ea[ea['Date'].dt.year >= 2020]
ea_latest = ea_latest[ea_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
microsoft_latest = microsoft[microsoft['Date'].dt.year >= 2020]
microsoft_latest = microsoft_latest[microsoft_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
take_two_latest = take_two[take_two['Date'].dt.year >= 2020]
take_two_latest = take_two_latest[take_two_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
ubisoft_latest = ubisoft[ubisoft['Date'].dt.year >= 2020]
ubisoft_latest = ubisoft_latest[ubisoft_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
zynga_latest = zynga[zynga['Date'].dt.year >= 2020]
zynga_latest = zynga_latest[zynga_latest['Date'].dt.day >= 22].reset_index().drop(columns = 'index')
ubisoft_latest
```

```python
#Filter all the date for covid cases from 2020-01-22 to 2022-4-29
lst = [i for i in covid.columns if ('20' in i) or ('21' in i) or ('22' in i)]
covid_data = covid[lst]
covid_data = covid_data.drop(columns = '5/5/22')
covid_data = covid_data.T
covid_data['cases'] = covid_data.sum(axis = 1)
#Sum up all the cases within the same date
covid_data = covid_data.reset_index()
covid_data = covid_data[['index','cases']]
covid_data.rename(columns = {'index':'Date'},inplace = True)
#Keep only the date and total cases
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
#convert date col into datetime object
covid_data.head()
```

```python
#Find Number of New Covid Cases Daily, which represent the severness of Covid-19
increase = np.diff(covid_data['cases'])
increase = np.insert(increase, 0,0)
covid_data['Increase']= increase
covid_data
```

Eventually, we would like to discover the relationship between stock price and covid cases. Let's use Nintendo as an example to do some inferential analysis.

### Nintendo and COVID-19 Cases

```python
nintendo_latest
```

```python
covid_data
```

```python
#merge nintendo stock price and covid cases
nintendo_data = pd.merge(nintendo_latest, covid_data, on=['Date'])
```

```python
nintendo_data
```

```python
#plot the relationship between nintendo stock price and covid test cases
sns.scatterplot(x='cases', y='Close', data=nintendo_data, s=100)
plt.xlabel('cases')
plt.ylabel('Close');
plt.title('Nintendo Model',fontsize = 30)
#sns.relplot(data=nintendo_data, x="cases", y="Close")
plt.xlabel('Cases',fontsize = 30)
plt.ylabel('Close',fontsize = 30)
```

From the graph generated above, the relationship is non-linear overall and do not meet the assumptions of linear regression. Thus, we would like to use a polynomial regression and see if that fits.

```python
import numpy as np
import matplotlib.pyplot as plt

x = nintendo_data['cases']
y = nintendo_data['Close']

#fit a polynomial of degree 3
model = np.poly1d(np.polyfit(x, y, 3))

poly_reg = np.linspace(1, 9e7, 100)

plt.scatter(x, y)
plt.plot(poly_reg, model(poly_reg))
plt.title('Nintendo Model',fontsize = 40)
plt.xlabel('Cases',fontsize = 30)
plt.ylabel('Close',fontsize = 30)
plt.show()
```

```python
#use MSE and coefficient of determination (R^2) to see how the model fits
from sklearn.metrics import mean_squared_error, r2_score
print('MSE:', mean_squared_error(y, model(x)))
print('R^2:', r2_score(y, model(x)))
```

The R^2 score is fairly high. We would like to use the model to predict Nintendo's stock price at a given covid cases number.

```python
#predict Nintendo's stock price at when the covid test cases is 6e7
price = model(6e7)
print(price)
```
